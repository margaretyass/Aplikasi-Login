### Pertanyaan 2:

```python
# Menghitung jumlah semua elemen pada kolom casual
total_casual = sum(day_df['casual'])

# Menghitung jumlah semua elemen pada kolom registered
total_registered = sum(day_df['registered'])

# Membuat data untuk pie plot
data = [total_casual, total_registered]
labels = ['Casual', 'Registered']

# Membuat pie plot
plt.pie(data, labels=labels, autopct='%1.1f%%', colors=["#FF7F50", "#A52A2A"])

# Menampilkan pie plot
plt.show()
```

**Insight:**
- Sebanyak 81,2% orang sudah registered dan sisanya 18,8% orang belum registered (casual)
